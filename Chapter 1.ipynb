```python
from sympy import *
from sympy.plotting import plot3d
```

```python
x = symbols('x')
f = 2*x + 1
plot(f)
```

```python
f2 = x**2 + 1
plot(f2)
```

```python
x, y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```

```python
summation = sum(2*i for i in range(1,6))
print(summation)
```

```python
x = [1,4,6,2]
n = len(x)

summation = sum(10*x[i] for i in range(0,n))
print(summation)
```

```python
i, n = symbols('i n')
summation = Sum(2*i, (i, 1, n))
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())
```

```python
x = symbols('x')
expr = x**2 / x**5
print(expr)
```
